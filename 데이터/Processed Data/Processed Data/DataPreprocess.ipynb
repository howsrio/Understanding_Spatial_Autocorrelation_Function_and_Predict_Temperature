```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
#AWS: 방재, ASOS: 종관
#온도 데이터 통합
Years=[2002,2007,2012,2017,2022]
Seasons=['Spring','Summer','Fall','Winter']
temps=[]
for i in Years:
    temps.append(pd.concat([pd.read_excel("C:/Users/bumju/Desktop/공간통계 학술제/raw_data/{}_Temp_ASOS.xlsx".format(i)),
    pd.read_excel("C:/Users/bumju/Desktop/공간통계 학술제/raw_data/{}_Temp_AWS.xlsx".format(i))]))
    

#월->계절 분류
def SeasonLabeling(df):
    data_seasoned=[]
    df['Season']=df['일시'].dt.month
    for i in range(1,13):
        if (i==3)|(i==4)|(i==5):
            df.loc[df['Season']==i,'Season']='Spring'
        elif (i==6)|(i==7)|(i==8):
            df.loc[df['Season']==i,'Season']='Summer'
        elif (i==9)|(i==10)|(i==11):
            df.loc[df['Season']==i,'Season']='Fall'
        elif (i==12)|(i==1)|(i==2):
            df.loc[df['Season']==i,'Season']='Winter'
    spring=df.loc[df['Season']=='Spring'].\
    pivot_table(index=['Point'],columns='Season',values='평균기온(°C)',aggfunc='mean')
    summer=df.loc[df['Season']=='Summer'].\
    pivot_table(index=['Point'],columns='Season',values='평균기온(°C)',aggfunc='mean')
    fall=df.loc[df['Season']=='Fall'].\
    pivot_table(index=['Point'],columns='Season',values='평균기온(°C)',aggfunc='mean')
    winter=df.loc[df['Season']=='Winter'].\
    pivot_table(index=['Point'],columns='Season',values='평균기온(°C)',aggfunc='mean')
    data_seasoned=[spring,summer,fall,winter]
    return data_seasoned
for i in range(0,5):
    temps[i].rename(columns={'지점명':'Point'},inplace=True)
    temps[i]=SeasonLabeling(temps[i])

#index[년도][계절]
```

```python
df_AWS_obs=pd.read_excel("C:/Users/bumju/Desktop/공간통계 학술제/raw_data/방재기상관측 관측지점 raw data.xlsx",header=1)
df_ASOS_obs=pd.read_excel("C:/Users/bumju/Desktop/공간통계 학술제/raw_data/종관기상관측 관측지점 raw data.xlsx",header=1)
df_total_obs=pd.concat([df_ASOS_obs,df_AWS_obs])
df_total_obs['ST_year']=df_total_obs['시작일'].dt.year
df_total_obs['ST_month']=df_total_obs['시작일'].dt.month
df_total_obs['END_year']=df_total_obs['종료일'].dt.year
df_total_obs['END_month']=df_total_obs['종료일'].dt.month
```

```python
df_total_obs.rename(columns={'위도':'Lat','경도':'Long','지점명':'Point'},inplace=True)
```

```python
def SeasonDataframe(year, month, df):
    df_season=df.loc[(((df['ST_year']<=year)&((year<df['END_year'])|(df['END_year'].isnull())))|
    ((df['END_year']==year)&(month<=df['END_month'])))].copy()
    return df_season
```

```python
obs=[]
for i in range(2002,2023,5):
    dfs=[]
    for j in range(5,12,3):
        dfs.append(SeasonDataframe(i,j,df_total_obs))
    dfs.append(SeasonDataframe(i+1,2,df_total_obs))
    obs.append(dfs)
```

```python
def DfMerge(df1,df2):
    try: df1.set_index('Point',inplace=True)
    except: pass
    df=df1.drop(['지점','시작일', '종료일', '지점주소', '관리관서','노장해발고도(m)',
       '기압계(관측장비지상높이(m))', '기온계(관측장비지상높이(m))', '풍속계(관측장비지상높이(m))',
       '강우계(관측장비지상높이(m))','ST_year','ST_month','END_year','END_month'], axis=1).join(df2,how='inner').copy()
    return df
```

```python
files=[]

for i in range(0,5):
    for j in range(0,4):
        DfMerge(obs[i][j],temps[i][j]).to_excel("C:/Users/bumju/Desktop/공간통계 학술제/processed_data/Temps_by_Obs_{}_{}.xlsx".format(Years[i],Seasons[j]))
```
